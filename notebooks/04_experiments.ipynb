# 04 — Experiments (Phase 4)

Reproducible validation experiments A–D from the article roadmap. Each figure is produced by a script under `scripts/` with fixed seeds.

| Experiment | Script | Figure |
|------------|--------|--------|
| A — Random annotators | `experiment_a_random.py` | `figures/exp_a_random_metrics.png` |
| B — High agreement trap | `experiment_b_trap.py` | `figures/exp_b_agreement_trap_heatmap.png` |
| C — LLM vs humans | `experiment_c_llm.py` | `figures/exp_c_llm_vs_humans.png` |
| D — Missing data | `experiment_d_missing.py` | `figures/exp_d_missing_robustness.png` |

From the repo root, run all at once: `make experiments` (or run the cell below).

```python
from __future__ import annotations

import subprocess
import sys
from pathlib import Path

_here = Path.cwd()
ROOT = _here.parent if _here.name == "notebooks" else _here

SCRIPTS = [
    ROOT / "scripts" / "experiment_a_random.py",
    ROOT / "scripts" / "experiment_b_trap.py",
    ROOT / "scripts" / "experiment_c_llm.py",
    ROOT / "scripts" / "experiment_d_missing.py",
]
for path in SCRIPTS:
    print("Running", path.name, "...")
    subprocess.run([sys.executable, str(path)], cwd=ROOT, check=True)
print("All experiments finished.")
```

```python
from pathlib import Path

from IPython.display import Image, display

_here = Path.cwd()
ROOT = _here.parent if _here.name == "notebooks" else _here

figs = [
    "exp_a_random_metrics.png",
    "exp_b_agreement_trap_heatmap.png",
    "exp_c_llm_vs_humans.png",
    "exp_d_missing_robustness.png",
]
for name in figs:
    display(Image(filename=str(ROOT / "figures" / name)))
```
