## Match the reference package (nominal)

```python
rel = np.array(
    [
        [np.nan, np.nan, np.nan, np.nan, np.nan, 3, 4, 1, 2, 1, 1, 3, 3, np.nan, 3],
        [1, np.nan, 2, 1, 3, 3, 4, 3, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan],
        [np.nan, np.nan, 2, 1, 3, 4, 4, np.nan, 2, 1, 1, 3, 3, np.nan, 4],
    ],
    dtype=float,
)
units_raters = rel.T
ours = krippendorff_alpha(units_raters, level_of_measurement="nominal")
ref = krippendorff.alpha(reliability_data=rel, level_of_measurement="nominal")
ours, ref, abs(ours - ref)
```
